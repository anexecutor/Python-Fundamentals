```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

x=np.linspace(-10,10,30)
noise=np.random.normal(0,5,x.shape)##normal表示正态分布,均值0,标准差5
y_true=x**2+3*np.sin(x)
y_pred=x**2+3*np.sin(x)+noise
```

```python
x_pred=np.linspace(x.min(),x.max(),300)

a=[1,3,9]
b=['red','green','blue']
for i in range(3):
    coef=np.polyfit(x,y,a[i])
    y_pred=np.polyval(coef,x_pred)
    
    plt.plot(x_pred,y_pred,color=b[i],label=f'Degree {a[i]}')

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Fits')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def mse(y_true,y_pred):
    return np.mean((y_true - y_pred)**2)

for i in range(3):
    print(f'Degree {a[i]} MSE:',mse(y_true,np.polyval(np.polyfit(x,y,a[i]),x)))
```

## 结论

- 增加模型复杂度会降低训练误差，但可能导致过拟合。
- 适中的多项式次数能够在偏差和方差之间取得最佳平衡。
