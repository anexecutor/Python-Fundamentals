#Experiment 03.1:模型欠拟合和过拟合

目的：研究模型复杂度对拟合程度的影响

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

x=np.linspace(-10,10,30)
noise=np.random.normal(0,0.5,x.shape)##normal表示正态分布,均值0,标准差5
y=5*np.sin(x) + x
y_obs=y+noise
```

```python
plt.scatter(x,y_obs,color='green',label='Noisy data')
plt.plot(x,y,color='red',label='True function')
plt.plot(x,y_obs,color='blue',label='Noisy data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Noisy Data')
plt.grid(True)##显示网格
plt.legend()
plt.show()
```

```python
x_pred=np.linspace(x.min(),x.max(),300)

a=[6,15,20]
b=['red','green','blue']
for i in range(3):
    coef=np.polyfit(x,y_obs,a[i])
    y_pred=np.polyval(coef,x_pred)
    
    plt.plot(x_pred,y_pred,color=b[i],label=f'Degree {a[i]}')

plt.scatter(x,y_obs)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Fits')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def mse(y_true,y_pred):
    return np.mean((y - y_pred)**2)

for i in range(3):
    print(f'Degree {a[i]} MSE:',mse(y,y_pred=np.polyval(np.polyfit(x,y_obs,a[i]),x)))
  
```
