# Experiment 01: Random Search for 1D Optimization

目标：寻找函数最小值  
函数：f(x)=x^2+3sin(x)

范围：[-10,10]  
方式：随机搜索

```python
import numpy as np##导入numpy库
import matplotlib.pyplot as plt##导入matplotlib.pyplot库
```

```python
def objective(x):##定义目标函数
    return x**2 + 3*np.sin(x)
```

```python
np.random.seed(0)##设置随机种子

N=100##定义样本数量，N表示生成1000个样本点
xs=np.random.uniform(-10,10,N)##在区间[-10,10]上均匀采样N个点
ys=objective(xs)##计算每个采样点对应的函数值
```

```python
best_index=np.argmin(ys)##找到函数值最小点的索引（下标），argmin函数返回最小值的索引
best_x=xs[best_index]##找到函数最小值点对应的自变量
best_y=ys[best_index]##找到函数最小值

best_x,best_y##输出最小值点的自变量和函数值
```

```python
x_plot=np.linspace(-10,10,100)##生成用于绘图的自变量
##linspace函数用于生成等间距的数值序列，此处为-10到10的100个点
y_plot=objective(x_plot)##计算绘图点对应的函数值

plt.figure(figsize=(10,6))##设置画布大小
plt.plot(x_plot,y_plot,color='black',label="f(x)")##绘制函数曲线

plt.scatter(xs,ys,s=1,alpha=0.3,label="samples")##绘制采样点
##s为点的大小，alpha为透明度，label为图例标签
plt.scatter(best_x,best_y,color='red',s=50,label="best point")

plt.legend()##显示图例（左下角）
plt.xlabel("x")##设置x轴标签
plt.ylabel("f(x)")##设置y轴标签
plt.title("Random Search for Data-driven Optimization")##设置图标标题
plt.show()##显示图像
```

## 结果分析

- 最小值对应的自变量值在-0.9139451250297963附近
- 最小值约为-1.5404609880722338
- 增加样本点数目能使结果更接近真实值
